### KNN from scratch

```python
class kNN():

    def fit(self, x, y):
        self.x = x
        self.y = y

    def euclidean_distance(self, x):
        # input: single data point
        if x.ndim == 1:
            l2 = np.sqrt(np.sum((self.x - x)**2, axis=1))

        # input: matrix of data points
        if x.ndim == 2:
            n_samples, _ = x.shape
            l2 = [np.sqrt(np.sum((self.x - x[i])**2, axis=1)) for i in range(n_samples)]

        return np.array(l2)

    def predict(self, x, k=1):

        # step 1: compute distance between input and training data
        dists = self.euclidean_distance(x)

        # step 2: find the k nearest neighbors and their classifications
        if x.ndim == 1:
            if k == 1:
                nn = np.argmin(dists)
                return self.y[nn]
            else:
                knn = np.argsort(dists)[:k]
                y_knn = self.y[knn]
                max_vote = max(y_knn, key=list(y_knn).count)
                return max_vote

        if x.ndim == 2:
            knn = np.argsort(dists)[:, :k]
            y_knn = self.y[knn]
            if k == 1:
                return y_knn.T[0]
            else:
                n_samples, _ = x.shape
                max_votes = [max(y_knn[i], key=list(y_knn[i]).count) for i in range(n_samples)]
                return max_votes
```

### Initializing and training the model

# Naive Bayes

```python
data = load_iris()
```

```python
x = pd.DataFrame(x, columns=column_names)
```

```python
x
```

```python
x.hist(bins=20)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=5)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
class MyNaiveBayes:
    def fit(self, x_train: pd.DataFrame, y_train: np.ndarray):
        self.x_train = x_train
        self.y_train = y_train
        
        self.means = self.x_train.groupby(self.y_train).mean()
        self.stds = self.x_train.groupby(self.y_train).apply(np.std)
        self.probs = self.x_train.groupby(self.y_train).apply(lambda x: len(x)) / self.x_train.shape[0]
    
    def predict(self, x_test: pd.DataFrame):
        # Вычисляем вероятность для Теоремы Байеса для каждого элемента
        y_pred = []
        self.y_pred_proba = []
        
        # каждый элемент в валидационной части данных
        for elem in range(x_test.shape[0]):
            p = {}

            # для каждого возможного класса
            for cl in np.unique(self.y_train):

                # априорная вероятность взятого ранее класса
                p[cl] = self.probs.iloc[cl]

                # для каждого столбца в датасете
                for index, param in enumerate(x_test.iloc[elem]):

                    # умножаем вероятность того, что данное значение столбца
                    # будет принадлежать распределению для выбранного класса
                    p[cl] *= norm.pdf(param, self.means.iloc[cl, index], self.stds.iloc[cl, index])
#                     p[cl] *= 1 / ( self.stds.iloc[cl, index] * np.sqrt(2 * np.pi) ) * np.exp( (-0.5) * (param - self.means.iloc[cl, index]) ** 2 / self.stds.iloc[cl, index] ** 2 )
   
            y_pred.append(pd.Series(p).values.argmax())
            self.y_pred_proba.append(pd.Series(p).values.max())
        
        return np.array(y_pred)
    
    def get_probs(self):
        if self.y_pred_proba:
            return np.array(self.y_pred_proba)
```

```python
nb = MyNaiveBayes()
```

```python
nb.fit(x_train, y_train)
```

```python
y_pred = nb.predict(x_test)
```

```python
nb.get_probs()
```

```python
y_pred
```

```python
accuracy1 = accuracy_score(y_test, y_pred)
accuracy1
```

### Плюсы

- Алгоритм легко и быстро предсказывает класс тестового набора данных. Он также хорошо справляется с многоклассовым прогнозированием.
- Производительность наивного байесовского классификатора лучше, чем у других простых алгоритмов, таких как логистическая регрессия. Более того, вам требуется меньше обучающих данных.
- Он хорошо работает с категориальными признаками(по сравнению с числовыми). Для числовых признаков предполагается нормальное распределение, что может быть серьезным допущением в точности нашего алгоритма.

### Минусы

- Если переменная имеет категорию (в тестовом наборе данных), которая не наблюдалась в обучающем наборе данных, то модель присвоит 0 (нулевую) вероятность и не сможет сделать предсказание. Это часто называют нулевой частотой. Чтобы решить эту проблему, мы можем использовать технику сглаживания. Один из самых простых методов сглаживания называется оценкой Лапласа.
- Значения спрогнозированных вероятностей, возвращенные методом predict_proba, не всегда являются достаточно точными.
- Ограничением данного алгоритма является предположение о независимости признаков. Однако в реальных задачах полностью независимые признаки встречаются крайне редко.

Метод Лапласа — метод, использующийся для приближённого вычисления интеграла вида


где $\phi(x)$ — некоторая дважды дифференцируемая функция, а $\lambda$ — некоторое большое число.

Предполагается, что функция $\phi(x)$ имеет единственный глобальный максимум в $x_0$. Тогда значение $\phi(x_0)$) будет большим, чем любое значение $\phi(x)$ в рассматриваемом промежутке интегрирования. Следовательно, для оценки этого интеграла можно ограничиться рассмотрением функции $\phi(x)$ лишь в небольшой окрестности глобального максимума. Для этого функции $\phi(x)$ и $\Phi(x)$ раскладывают в ряд Тейлора в окрестности этой точки.

### В каких областях использовать?

<b>Алгоритм наивного Байеса – это классификатор, обучение которого идет очень быстро. Следовательно, данный инструмент идеально подходит для составления прогнозов в реальном времени.</b>

Также этот алгоритм также хорошо известен благодаря функции многоклассового прогнозирования. Мы можем предсказать вероятность нескольких классов целевой переменной.

Таким образом, идеальные области для применения наивного байесовского классификатора это:

- <b>Система рекомендаций</b>. В сочетании алгоритма с методами <a href = 'https://ru.wikipedia.org/wiki/%D0%9A%D0%BE%D0%BB%D0%BB%D0%B0%D0%B1%D0%BE%D1%80%D0%B0%D1%82%D0%B8%D0%B2%D0%BD%D0%B0%D1%8F_%D1%84%D0%B8%D0%BB%D1%8C%D1%82%D1%80%D0%B0%D1%86%D0%B8%D1%8F'>коллаборативной фильтрации</a> (Collaborative Filtering) мы можем создать рекомендательную систему, которая использует машинное обучение и методы добычи данных для учета невидимой информации (такой, как поиск фильмов пользователем и длительность просмотра). Цель – предсказание того, понравится ли пользователю данный ресурс/продукт или нет.
- <b>Фильтрация спама и классификация текста</b>. Наивный байесовский классификатор в основном используются для классификации текстов (благодаря лучшему результату в многоклассовых проблемах и правилу независимости) и имеет более высокую точность по сравнению с другими алгоритмами. В результате он широко используется в фильтрации спама (в электронной почте) и анализе настроений (к примеру, социальных сетей, для выявления положительных и отрицательных настроений клиентов).
